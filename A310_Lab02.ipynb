### **Problem 1**

```python
# define a sample space for a fair, unbiased coin;
sample_space = {'Heads', 'Tails'}

# define a weighted sample space for a biased coin that generates tails three times more often than heads;
weighted_sample_space = {'Heads': 1, 'Tails': 3}

# write a function compute_event_probability that calculates the probability of each elementary event (heads, tails, either, or neither) in both cases.
# bring in the event helper from the textbook
def get_matching_event(event_condition, sample_space):
    return set([outcome for outcome in sample_space
                if event_condition(outcome)])

#main event_condition function
def compute_event_probability(event_condition, generic_sample_space):
    event = get_matching_event(event_condition, generic_sample_space)
    if type(generic_sample_space) == type(set()):
        return len(event) / len(generic_sample_space)

    # weighted event condition function
    event_size = sum(generic_sample_space[outcome] for outcome in event)
    return event_size / sum(generic_sample_space.values())
```

### **Problem 2**

```python
# What is the probability that exactly two of the children are girls?
# creates a sample space using product from itertools
from itertools import product
possible_children = ['Boy', 'Girl']
sample_space = set(product(possible_children, repeat=4))

# defining the event condition = exactly two boys
def has_two_boys(outcome):
  return len([child for child in outcome if child == 'Boy']) == 2

# calculate the probability of the event condition
prob = compute_event_probability(has_two_boys, sample_space)
print(f'Probability of having exactly 2 boys is {prob}')
```

### **Problem 3**

```python
# eight diew rolls summing to 28
# importing the library needed for this problem
from collections import defaultdict

# define possible die rolls
possible_rolls = list(range(1, 7))

# creating the unweighted sample space for 8 rolls
sample_space = set(product(possible_rolls, repeat=8))

# convert to weighted sample space (map sums to occurrence counts)
weighted_sample_space = defaultdict(int)
for outcome in sample_space:
    total = sum(outcome)
    weighted_sample_space[total] += 1

# converting defaultdict to regular dictionary
weighted_sample_space = dict(weighted_sample_space)

# define the event condition = sum equals 28
def has_sum_of_28(outcome):
  return outcome == 28

#calculate probability
prob = compute_event_probability(has_sum_of_28, weighted_sample_space)
print(f'Probability of 8 rolls summing to 28 is {prob}')
```

### **Problem 4**

```python
# compute the likelihood that our eight consecutive die rolls sum to a value between 14 and 24 inclusive
# define our interval function
def is_in_interval(number, minimum, maximum):
    return minimum <= number <= maximum

# calculate probabaility using lambda expression
prob = compute_event_probability(lambda x: is_in_interval(x, 14, 24),weighted_sample_space)
print(f'Probability of interval is {prob}')
```

### **Problem 5**

```python
#import default libraries
from collections import defaultdict
from itertools import product

# create a function to generate weighted sample space for coin flips
def generate_coin_sample_space(num_flips=10):
  weighted_sample_space = defaultdict(int)
  for outcome in product(['Heads', 'Tails'], repeat=num_flips):
    heads_count = len([flip for flip in outcome if flip == 'Heads'])
    weighted_sample_space[heads_count] += 1
  return dict(weighted_sample_space)

# what is the probability that 10 fair coin flips produce either 0 to 2 heads or 8 to 10 heads?
weighted_sample_space_10_flips = generate_coin_sample_space(num_flips=10)
prob = compute_event_probability(lambda x: not is_in_interval(x, 3, 7), weighted_sample_space_10_flips)
print(f'Probability of observing extreme head counts (0-2 or 8-10) in 10 flips is {prob}')

# determine the probability of 20 fair coin fips not producing from 5 to 15 heads (or tails).
weighted_sample_space_20_flips = generate_coin_sample_space(num_flips=20)
prob = compute_event_probability(lambda x: not is_in_interval(x, 5, 15), weighted_sample_space_20_flips)
print(f'Probability of observing extreme head counts (0-4 or 16-20) in 20 flips is {prob}')
```
